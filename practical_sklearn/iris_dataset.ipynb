```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier   #try different model also like from sklearn.naive_bayes import GaussianNB
data=pd.read_excel("iris_train.xlsx")
data.head()
```

```python
x=data.iloc[:, :-1]  #take all columns except the last one
```

```python
x
```

```python
y=data.iloc[: , -1]
```

```python
y
```

```python
#training the model
model=DecisionTreeClassifier()   #make changes here according to the model used.
model.fit(x.values,y.values)
```

```python
predictions=model.predict([[6.8,2.8,4.8,1.4]])  
```

```python
predictions
```
